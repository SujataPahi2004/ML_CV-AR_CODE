```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load image as NumPy array (from OpenCV)
def load_image(image_path):
    return cv2.imread(image_path)

# Convert image to grayscale (without using any libraries)
def convert_grayscale(image):
    # Use the standard formula: Gray = 0.299 * R + 0.587 * G + 0.114 * B
    gray = 0.299 * image[:, :, 2] + 0.587 * image[:, :, 1] + 0.114 * image[:, :, 0]
    return gray.astype(np.uint8)

# Display image using OpenCV
def display_image(window_name, image):
    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Convert image to grayscale using OpenCV
def convert_grayscale_cv(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Example usage
image_path = '/content/n.c. college_logo.jpg'  # Replace with your image path
image = load_image(image_path)

# Manual grayscale conversion
gray_image_manual = convert_grayscale(image)
display_image("Manual Grayscale", gray_image_manual)

# OpenCV grayscale conversion
gray_image_cv = convert_grayscale_cv(image)
display_image("OpenCV Grayscale", gray_image_cv)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load image as NumPy array (from OpenCV)
def load_image(image_path):
    return cv2.imread(image_path)

# Convert image to redscale
def convert_to_redscale(image):
    # Create an empty array with the same shape as the input image
    redscale_image = np.zeros_like(image)
    # Copy the red channel values into the new image
    redscale_image[:, :, 2] = image[:, :, 2]  # R channel
    return redscale_image

# Display image using OpenCV
def display_image(window_name, image):
    cv2_imshow(image)

# Example usage
image_path = '/content/n.c. college_logo.jpg'  # Replace with your image path
image = load_image(image_path)

# Convert to redscale
redscale_image = convert_to_redscale(image)
display_image("Redscale Image", redscale_image)
```
